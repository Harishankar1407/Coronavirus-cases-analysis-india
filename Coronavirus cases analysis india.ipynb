```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
mpl.get_backend()
```

```python
df=pd.read_excel(r'C:\Users\Asus\Downloads\Book 1.xlsx')
```

```python
df
```

```python
df['Female(%)']=df['Female(%)'].replace([62,68,66,64,58,64,75],[38,32,33,36,42,36,25])
```

```python
df
```

```python
df.set_index('State')
```

```python
x=df['Male(%)']
y=df['Female(%)']
z=df['State']
plt.figure()
plt.plot(z,x,'-o',z,y,'-o')
```

```python
plt.title('Coronavirus cases in states on the basis of gender')
plt.legend(['Male','Female'],frameon=False)
```

```python
plt.xlabel('States')
plt.ylabel('Percentage')
```

```python
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
```

```python
plt.subplots_adjust(bottom=0.25)
```

```python
df.plot(kind="barh", width=0.8)
for i, (p, pr) in enumerate(zip(df['State'], df['Male(%)'])):
    plt.text(s=p, x=1, y=i, color="w", verticalalignment="center", size=10)
    plt.text(s=str(pr)+"%", x=pr-5, y=i, color="black",
             verticalalignment="center", horizontalalignment="left", size=16)
for i, (p, pr) in enumerate(zip(df['State'], df['Female(%)'])):
    plt.text(s=p, x=1, y=i, color="w", verticalalignment="center", size=10)
    plt.text(s=str(pr)+"%", x=pr-5, y=i, color="black",
             verticalalignment="center", horizontalalignment="left", size=16,clip_on=True)
plt.axis("off")

plt.legend(loc=4)
plt.title('Coronavirus cases in states of india categorized by gender')
plt.tight_layout()
```
